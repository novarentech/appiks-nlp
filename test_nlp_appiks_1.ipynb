```python
import re

def tokenize(text):
    # Lower case dulu
    text = text.lower()
    # Hapus tanda baca, ganti dengan spasi
    text = re.sub(r'[^a-z0-9\\s]', ' ', text)
    # Split jadi tokens, hapus token kosong
    tokens = [t for t in text.split() if t]
    return tokens

# Test
teks = "Aku Sedih. Tidak tahu mau apa lagi!"
tokens = tokenize(teks)
print(f"Teks asli: {teks}")
print(f"Tokens: {tokens}")
```

Stopword Removal

```python
STOPWORDS_BASIC = {
    'aku', 'saya', 'kamu', 'dia', 'kita',
    'dan', 'yang', 'di', 'ke', 'dari', 'untuk', 'pada',
    'adalah', 'ini', 'itu', 'tersebut',
    'sangat', 'sekali', 'banget',
    'ya', 'sih', 'dong', 'kok', 'nih'
}

def remove_stopwords(tokens):
    return [t for t in tokens if t not in STOPWORDS_BASIC]

# Test
tokens = ['aku', 'merasa', 'sedih', 'banget', 'hari', 'ini']
tokens_clean = remove_stopwords(tokens)

print(f"Sebelum: {tokens}")
print(f"Sesudah: {tokens_clean}")
print(f"\\nYang dibuang: {set(tokens) - set(tokens_clean)}")
```

```python
def preprocess_appiks(text):
  text = text.lower()
  text = re.sub(r'[^a-z0-9\\s]', ' ', text)
  text = stopword_remover.remove(text)
  stemmed_text = stemmer.stem(text)
  tokens = [t for t in stemmed_text.split() if t]
  return tokens

kalimat_test = [
    "Aku merasa sangat kesepian dan tidak berguna.",
    "Hari ini capek banget, ga ada gunanya hidup gini.",
    "Pengen mati aja rasanya, ga ada yang peduli.",
    "Aku tidak tahu harus berbuat apa lagi."
]

for k in kalimat_test:
    print(f"Input : {k}")
    print(f"Stems : {preprocess_appiks(k)}")
    print()
```

## Keyword Matching

```python
KAMUS_RED_NAIVE = {"bunuh", "mati", "sayat", "akhir"}
KAMUS_YELLOW_NAIVE = {"sedih", "capek", "kosong", "hampa", "sendiri", "guna"}

def classify_naive(text):
    tokens = preprocess_appiks(text)

    # Cek Red Zone dulu
    for token in tokens:
        if token in KAMUS_RED_NAIVE:
            return "Red Zone", [token]

    # Cek Yellow Zone
    matched = [t for t in tokens if t in KAMUS_YELLOW_NAIVE]
    if matched:
        return "Yellow Zone", matched

    return "No Trigger", []

# Test
test_cases = [
    "Aku merasa sangat kesepian dan kosong",
    "Aku ingin akhiri hidup ini",
    "Hari ini biasa-biasa saja",
    "Capek banget belajar matematika"
]

for tc in test_cases:
    zone, matched = classify_naive(tc)
    print(f"Teks: {tc}")
    print(f"  → Zone: {zone}, Matched: {matched}")
    print()
```

```python
problematic_cases = [
    # Kasus 1: kata tunggal yang generic
    ("Hari ini capek banget belajar matematika", "Bukan distress, hanya keluhan akademik"),

    # Kasus 2: negasi
    ("Aku tidak ingin bunuh diri", "Sebenarnya MENGAKU tidak mau"),

    # Kasus 3: kata generic dalam konteks lain
    ("Lemari kosong, lupa belanja", "Bukan distress psikologis"),

    # Kasus 4: kata generic
    ("Aku capek karena olahraga tadi", "Capek fisik, bukan distress"),
]

print("=" * 70)
print("DEMONSTRASI KELEMAHAN PENDEKATAN NAIVE")
print("=" * 70)
for teks, realitas in problematic_cases:
    zone, matched = classify_naive(teks)
    status = "SALAH" if zone != "No Trigger" else "✓ Benar"
    print(f"\\nTeks: '{teks}'")
    print(f"Realitas: {realitas}")
    print(f"Hasil Naive: {zone} (matched: {matched}) {status}")
```

### **Weighted Scoring 3-Lapis**

Setiap keyword punya bobot 1-10 ->
Klasifikasi berdasarkan total skor (bukan keberadaan keyword) ->
Bonus +2 untuk kombinasi keyword yang klinis bermakna


```python
KAMUS_BERBOBOT = {
    # ===== RED ZONE =====
    "bunuh": (10, "Red"),
    "akhir": (9, "Red"),   # "akhiri hidup"
    "sayat": (9, "Red"),
    "luka": (9, "Red"),    # "lukai diri"
    "gantung": (10, "Red"),
    "racun": (8, "Red"),
    "mati": (7, "Red"),

    # ===== YELLOW ZONE - bobot tinggi (distress kuat) =====
    "guna": (5, "Yellow"),    # "ga ada gunanya"
    "peduli": (5, "Yellow"),  # "ga ada yang peduli"
    "harap": (6, "Yellow"),   # "ga ada harapan"
    "nyerah": (5, "Yellow"),
    "beban": (6, "Yellow"),   # "beban orang"

    # ===== YELLOW ZONE - bobot sedang =====
    "kosong": (4, "Yellow"),
    "hampa": (4, "Yellow"),
    "sendiri": (3, "Yellow"),
    "tangis": (5, "Yellow"),  # "menangis terus"

    # ===== YELLOW ZONE - bobot rendah (butuh konteks) =====
    "capek": (1, "Yellow"),
    "lelah": (1, "Yellow"),
    "malas": (1, "Yellow"),
    "bosan": (1, "Yellow"),
}

# Co-occurrence pairs - kombinasi bermakna klinis
CO_OCCURRENCE_PAIRS = [
    ("capek", "guna"),        # anhedonia + hopelessness
    ("sendiri", "peduli"),    # loneliness + abandonment
    ("kosong", "harap"),      # emptiness + hopelessness
    ("lelah", "beban"),       # fatigue + burden
]

# Threshold per zona
THRESHOLD_YELLOW = 3
THRESHOLD_RED = 10

print(f"Kamus berbobot: {len(KAMUS_BERBOBOT)} keyword")
print(f"Co-occurrence pairs: {len(CO_OCCURRENCE_PAIRS)} pasang")
```

```python
def classify_weighted(text):
    # Step 1: Preprocess
    tokens = preprocess_appiks(text)

    # Step 2: Match keyword + hitung weighted score
    matched_keywords = []
    base_score = 0
    has_red_explicit = False

    for token in tokens:
        if token in KAMUS_BERBOBOT:
            weight, zone = KAMUS_BERBOBOT[token]
            matched_keywords.append({
                "stem": token,
                "weight": weight,
                "zone": zone
            })
            base_score += weight

            # Lapis 2 override: kata kunci Red Zone yang sangat eksplisit
            if zone == "Red" and weight >= 9:
                has_red_explicit = True

    # Step 3: Co-occurrence Bonus (Lapis 3)
    matched_stems = {m["stem"] for m in matched_keywords}
    bonus_score = 0
    bonus_pairs = []

    for stem1, stem2 in CO_OCCURRENCE_PAIRS:
        if stem1 in matched_stems and stem2 in matched_stems:
            bonus_score += 2
            bonus_pairs.append((stem1, stem2))

    total_score = base_score + bonus_score

    # Step 4: Klasifikasi zona (Lapis 2 threshold)
    if has_red_explicit or total_score >= THRESHOLD_RED:
        zone = "Red Zone"
    elif total_score >= THRESHOLD_YELLOW:
        zone = "Yellow Zone"
    else:
        zone = "No Trigger"

    # Breakdown untuk audit & explainability
    breakdown = {
        "base_score": base_score,
        "bonus_score": bonus_score,
        "bonus_pairs": bonus_pairs,
        "total_score": total_score,
        "has_red_explicit": has_red_explicit,
        "threshold_yellow": THRESHOLD_YELLOW,
        "threshold_red": THRESHOLD_RED
    }

    return zone, matched_keywords, total_score, breakdown
```

```python
problematic_cases = [
    ("Hari ini capek banget belajar matematika", "No Trigger (bukan distress)"),
    ("Aku capek karena olahraga tadi", "No Trigger (capek fisik)"),
    ("Aku merasa sangat kesepian dan kosong", "Yellow Zone (distress)"),
    ("Capek, ga ada gunanya hidup gini, sendirian terus", "Red Zone (multiple keyword)"),
    ("Aku ingin akhiri hidup ini", "Red Zone (explicit)"),
    ("Pengen mati aja", "Red Zone (death wish)"),
]

print("=" * 80)
print("PERBANDINGAN: NAIVE vs WEIGHTED SCORING")
print("=" * 80)

for teks, ekspektasi in problematic_cases:
    # Versi naive
    zone_naive, _ = classify_naive(teks)

    # Versi weighted
    zone_weighted, matched, score, breakdown = classify_weighted(teks)

    print(f"\n Teks: '{teks}'")
    print(f"   Ekspektasi : {ekspektasi}")
    print(f"   Naive      : {zone_naive}")
    print(f"   Weighted   : {zone_weighted} (skor={score}, base={breakdown['base_score']}, bonus={breakdown['bonus_score']})")

    if matched:
        keywords_info = [f"{m['stem']}({m['weight']})" for m in matched]
        print(f"   Keywords matched: {', '.join(keywords_info)}")

    if breakdown['bonus_pairs']:
        print(f"   Co-occurrence bonus: {breakdown['bonus_pairs']}")
```

```python
import pandas as pd

labels = ["No Trigger", "Yellow Zone", "Red Zone"]
confusion = pd.DataFrame(
    [[0, 0, 0], [0, 0, 0], [0, 0, 0]],
    index=labels,
    columns=labels
)

for h in hasil_evaluasi:
    confusion.loc[h["expected"], h["predicted"]] += 1

# Tampilkan
print("CONFUSION MATRIX")
print("(baris=Expected, kolom=Predicted)")
print()
print(confusion)
```

```python
def calculate_metrics(confusion_matrix, target_class):
    tp = confusion_matrix.loc[target_class, target_class]

    # FP: prediksi target_class tapi expected bukan target_class
    fp = confusion_matrix[target_class].sum() - tp

    # FN: expected target_class tapi prediksi bukan target_class
    fn = confusion_matrix.loc[target_class].sum() - tp

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

    return precision, recall, f1, tp, fp, fn

print(f"{'Zone':<15} {'Precision':<12} {'Recall':<10} {'F1-Score':<10} {'TP':<5} {'FP':<5} {'FN':<5}")
print("-" * 75)
for zone in labels:
    p, r, f1, tp, fp, fn = calculate_metrics(confusion, zone)
    print(f"{zone:<15} {p:.2%}        {r:.2%}     {f1:.2%}     {tp:<5} {fp:<5} {fn:<5}")
```

```python
errors = [h for h in hasil_evaluasi if not h["correct"]]

print(f"📊 Total kesalahan: {len(errors)} dari {len(hasil_evaluasi)}")
print()
print("DAFTAR ERRORS:")
print("=" * 80)

for i, e in enumerate(errors, 1):
    # Tipe error
    if e["expected"] == "No Trigger" and e["predicted"] != "No Trigger":
        error_type = "FALSE POSITIVE"
    elif e["expected"] != "No Trigger" and e["predicted"] == "No Trigger":
        error_type = "FALSE NEGATIVE"
    else:
        error_type = "WRONG ZONE"

    print(f"\\n{i}. {error_type}")
    print(f"   Teks      : {e['teks']}")
    print(f"   Expected  : {e['expected']}")
    print(f"   Predicted : {e['predicted']}")
    print(f"   Keywords  : {e['matched']}")
    print(f"   Skor      : {e['score']}")
```
